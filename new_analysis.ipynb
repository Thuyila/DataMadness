```python
import pandas as pd
```

```python
binance = pd.read_csv('Binance_Coin_data.csv')
bitcoin = pd.read_csv('bitcoin_data.csv')
cardano = pd.read_csv('Cardano_data.csv')
chainLink = pd.read_csv('ChainLink_data.csv')
ethereum = pd.read_csv('Ethereum_data.csv')
litecoin = pd.read_csv('Litecoin_data.csv')
polkadot = pd.read_csv('Polkadot_data.csv')
stellar = pd.read_csv('Stellar.csv')
tether = pd.read_csv('Tether_data.csv')
xrp = pd.read_csv('XRP_data.csv')
```

```python
indices = [binance, bitcoin, cardano, chainLink, ethereum, litecoin, polkadot, stellar, tether, xrp]
names = ['Binance', 'Bitcoin', 'Cardano', 'ChainLink', 'Ethereum', 'Litecoin', 'Polkadot', 'Stellar', 'Tether', 'XRP']
binance['Crypto'] = binance['Open']
bitcoin['Crypto'] = bitcoin['Open']
cardano['Crypto'] = cardano['Open']
chainLink['Crypto'] = chainLink['Open']
ethereum['Crypto'] = ethereum['Open']
litecoin['Crypto'] = litecoin['Open']
polkadot['Crypto'] = polkadot['Open']
stellar['Crypto'] = stellar['Open']
tether['Crypto'] = tether['Open']
xrp['Crypto'] = xrp['Open']
```

```python
j = 0
for col in indices:
    index = []
    for i in range(len(binance)):
        index.append(names[j])
    col['Crypto'] = names[j]
    j += 1
```

```python
binance = binance[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
bitcoin = bitcoin[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
cardano = cardano[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
chainLink = chainLink[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
ethereum = ethereum[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
litecoin = litecoin[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
polkadot = polkadot[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
stellar = stellar[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
tether = tether[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
xrp = xrp[['Crypto', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
```

```python
dataset = pd.concat([binance, bitcoin, cardano, chainLink, ethereum, litecoin, polkadot, stellar, tether, xrp])
```
