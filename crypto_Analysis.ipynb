```python
import pandas as pd
```

```python
# import all datasets
binance = pd.read_csv('Binance_Coin_data.csv')
bitcoin = pd.read_csv('bitcoin_data.csv')
cardano = pd.read_csv('Cardano_data.csv')
chainLink = pd.read_csv('ChainLink_data.csv')
ethereum = pd.read_csv('Ethereum_data.csv')
litecoin = pd.read_csv('Litecoin_data.csv')
polkadot = pd.read_csv('Polkadot_data.csv')
stellar = pd.read_csv('Stellar.csv')
tether = pd.read_csv('Tether_data.csv')
xrp = pd.read_csv('XRP_data.csv')
```

```python
binance.head()
```

```python
# rename all column names
binance_new = binance.rename(columns = { 'Open' : 'binance_open', 'High': 'binance_high', 'Low':'binance_low', 'Close' : 'binance_close','Volume': 'binance_volume', 'Market Cap' : 'binance_MarketCap'})

bitcoin_new = bitcoin.rename(columns = { 'Open' : 'bitcoin_open', 'High': 'bitcoin_high', 'Low':'bitcoin_low', 'Close' : 'bitcoin_close','Volume': 'bitcoin_volume', 'Market Cap' : 'bitcoin_MarketCap'})

cardano_new = cardano.rename(columns = { 'Open' : 'cardano_open', 'High': 'cardano_high', 'Low':'cardano_low', 'Close' : 'cardano_close','Volume': 'cardano_volume', 'Market Cap' : 'cardano_MarketCap'})

chainLink_new = chainLink.rename(columns = { 'Open' : 'chainLink_open', 'High': 'chainLink_high', 'Low':'chainLink_low', 'Close' : 'chainLink_close','Volume': 'chainLink_volume', 'Market Cap' : 'chainLink_MarketCap'})

ethereum_new = ethereum.rename(columns = { 'Open' : 'ethereum_open', 'High': 'ethereum_high', 'Low':'ethereum_low', 'Close' : 'ethereum_close','Volume': 'ethereum_volume', 'Market Cap' : 'ethereum_MarketCap'})

litecoin_new = litecoin.rename(columns = { 'Open' : 'litecoin_open', 'High': 'litecoin_high', 'Low':'litecoin_low', 'Close' : 'litecoin_close','Volume': 'litecoin_volume', 'Market Cap' : 'litecoin_MarketCap'})

polkadot_new = polkadot.rename(columns = { 'Open' : 'polkadot_open', 'High': 'polkadot_high', 'Low':'polkadot_low', 'Close' : 'polkadot_close','Volume': 'polkadot_volume', 'Market Cap' : 'polkadot_MarketCap'})

stellar_new = stellar.rename(columns = { 'Open' : 'stellar_open', 'High': 'stellar_high', 'Low':'stellar_low', 'Close' : 'stellar_close','Volume': 'stellar_volume', 'Market Cap' : 'stellar_MarketCap'})

tether_new = tether.rename(columns = { 'Open' : 'tether_open', 'High': 'tether_high', 'Low':'tether_low', 'Close' : 'tether_close','Volume': 'tether_volume', 'Market Cap' : 'tether_MarketCap'})

xrp_new = xrp.rename(columns = { 'Open' : 'xrp_open', 'High': 'xrp_high', 'Low':'xrp_low', 'Close' : 'xrp_close','Volume': 'xrp_volume', 'Market Cap' : 'xrp_MarketCap'})
```

```python
# size of datasets
print('shape of binance',binance_new.shape)
print('shape of bitcoin',bitcoin_new.shape)
print('shape of cardano',cardano_new.shape)
print('shape of chainLink',chainLink_new.shape)
print('shape of ethereum',ethereum_new.shape)
print('shape of litecoin',litecoin_new.shape)
print('shape of polkadot',polkadot_new.shape)
print('shape of stellar',stellar_new.shape)
print('shape of tether',tether_new.shape)
print('shape of xrp',xrp_new.shape)
```

```python
# merge all datasets
dataset = pd.merge(cardano_new,polkadot_new,on = ['Date'], how = 'left')
dataset = pd.merge(dataset,chainLink_new,on=['Date'], how= 'right')
dataset = pd.merge(dataset,binance_new,on= ['Date'], how = 'right')
dataset = pd.merge(dataset,ethereum_new,on= ['Date'], how = 'right')
dataset = pd.merge(dataset,tether_new,on=['Date'], how = 'right' )
dataset = pd.merge(dataset,stellar_new,on=['Date'], how= 'right')
dataset = pd.merge(dataset,xrp_new,on=['Date'], how = 'right')
dataset = pd.merge(dataset, litecoin_new, on = ['Date'], how = 'right')
dataset = pd.merge(dataset, bitcoin_new, on= ['Date'], how = 'right')
dataset
```

```python
dataset.to_csv('crypto_dataset.csv')
```
